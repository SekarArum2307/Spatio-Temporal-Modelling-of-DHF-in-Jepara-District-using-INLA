# üñä **Standardized Incidence Ratio Calculation**

 Sometimes we wish to provide disease risk estimates in each of the areas that
 form a partition of the study region (Moraga, 2018). One simple measure
 of disease risk is the standardized incidence ratio (SIR). For each area i,
 i = 1,...,n, the SIR is defined as the ratio of observed counts to the expected
 counts

>  ***SIRi = Yi/Ei***

>The **expected counts** in each village represent the total number of disease cases one would expect if the population in the village behaved the way the population of Jepara behaves. We can do this by using the expected() function of **SpatialEpi**. This function has three arguments, namely,
*  **population**: vector of population counts for each strata in each area,
*  **cases**: vector with the number of cases for each strata in each area,
*  **n.strata**: number of strata (specified groups of gender, age, and race)


üåç*Moraga, P. (2019). Geospatial Health Data: Modeling and Visualization with R-INLA and Shiny* Chapman and Hall/CRC. https://www.crcpress.com/Geospatial-Health-Data-Modeling-and-Visualization-with-R-INLA-and-Shiny/Moraga/p/book/9780367357955.



## Expected Counts

```r
# Menghitung expected case
E <- expected(
  population = df$Populasi,
  cases = df$DBD, n.strata = 14
)
df$E <- E[match(df$ID_DESA, unique(df$ID_DESA))]
summary(df$E)
```

## SIR

```r
# Menghitung SIR
df$SIR <- df$DBD / df$E
summary(df$SIR)
```

## Standardized Scaling

Pada teknik standardized scaling, standarisasi kumpulan data melibatkan penskalaan ulang distribusi nilai sehingga rata-rata dari nilai yang diamati adalah 0 dan simpangan bakunya adalah 1. Hal ini dapat dianggap sebagai pengurangan nilai rata-rata atau memusatkan data. Seperti normalisasi, standardisasi dapat berguna, dan bahkan diperlukan dalam beberapa pemodelan ketika data memiliki nilai masukan dengan skala yang berbeda (Brownlee, 2020).

## ‚ñ∂ *Spatio-Temporal Model by Paula Moraga*

> Model spatio-temporal menggunakan distribusi Poisson untuk menganalisis risiko penyakit dengan mempertimbangkan efek spasial terstruktur dan tidak terstruktur, serta interaksi spasio-temporal . Formula modelnya menyatakan bahwa log dari risiko relatif Œ∏ij adalah fungsi dari intercept Œ±, efek spasial ui + ùë£i , dan interaksi waktu-area ((Œ≤+Œ¥i)√ótj).



üåç*Moraga, P. (2019). Geospatial Health Data: Modeling and Visualization with R-INLA and Shiny* Chapman and Hall/CRC. https://www.crcpress.com/Geospatial-Health-Data-Modeling-and-Visualization-with-R-INLA-and-Shiny/Moraga/p/book/9780367357955.



### Model 1 : Poisson Model

```r
# Definisikan formula model
formula_poisson <- O ~ temperature + rainfall+humidity+
  f(id.zone, model = "bym2", graph = jpr.adj, scale.model = TRUE) +
  f(id.zone.int, id.month, model = "iid") +
  id.month

# Terapkan formula model pada data
res_poisson <- inla(formula_poisson,
            data = data.INLA_scaled,
            control.predictor = list(link = 1, compute = TRUE),
            control.compute = list(dic = TRUE, waic = TRUE, cpo = TRUE),
            control.fixed = list(mean = 0, prec = 0.00001),
            family = "poisson"
)
```

```r
summary(res_poisson)
```

```r
summary(res_poisson$summary.fitted.values["mean"])
```

```r
summary(res_poisson$cpo$failure)
```

### Cross- validation

```r
dengue_data <- data.INLA_scaled
set.seed(123)  # Untuk reproduktifitas
fold_id <- sample(rep(1:10, length.out = nrow(dengue_data)))
```

```r
# Inisialisasi hasil
MAE_list_mod1 <- numeric(10)
RMSE_list_mod1 <- numeric(10)

for (fold in 1:10) {
  # Data untuk fold saat ini
  test_data <- dengue_data[fold_id == fold, ]
  train_data <- dengue_data[fold_id != fold, ]

  # Melatih model pada data pelatihan
  mod1 <- inla(O ~ temperature + rainfall+humidity+
  f(id.zone, model = "bym2", graph = jpr.adj, scale.model = TRUE) +
  f(id.zone.int, id.month, model = "iid") +
  id.month, data = train_data, control.predictor = list(link = 1, compute = TRUE),
  control.compute = list(dic = TRUE, waic = TRUE, cpo = TRUE),
  control.fixed = list(mean = 0, prec = 0.00001),
  family = "poisson")

  # Prediksi pada data uji
  expected_cases_test <- test_data$E
  predicted_cases <- mod1$summary.fitted.values$mean

  # Menghitung metrik evaluasi
  observed_cases_test <- test_data$O
  MAE <- mean(abs(observed_cases_test - predicted_cases))
  RMSE <- sqrt(mean((observed_cases_test - predicted_cases)^2))

  # Simpan hasil
  MAE_list_mod1[fold] <- MAE
  RMSE_list_mod1[fold] <- RMSE
}
```

```r
# Ringkasan hasil cross-validation
cat("Mean MAE:", mean(MAE_list_mod1), "\n")
cat("Mean RMSE:", mean(RMSE_list_mod1), "\n")
```

```r
library(INLA)

# Misalkan 'result' adalah objek dari model INLA
posterior_means <- res_poisson$summary.fitted.values$mean
observed_values <- data.INLA$O  # Menggunakan nilai pengamatan sebenarnya

# Hitung varians prediksi dan varians residual
var_y_hat <- var(posterior_means)
residuals <- observed_values - posterior_means
var_residuals <- var(residuals)

# Hitung R-squared
R2_mod1 <- var_y_hat / (var_y_hat + var_residuals)
print(R2_mod1)
```
